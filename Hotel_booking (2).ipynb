## 3. ***Data Wrangling***

### Data Wrangling Code

````python
#null values per column
df.isnull().sum()
````

````python
#storing the number of missing values in descending order
missing_val = df.isnull().sum().sort_values(ascending=False)
missing_val
````

````python
#percentage of null value in column company
null_prcnt_company = missing_val[0]/ unique_rows *100
null_prcnt_company
````

````python
#Percentage of missing values in column company is high so it is better to drop the column
df.drop(['company'], axis = 1, inplace=True)
````

````python
#percentage of null value in column agent
null_prcnt_agent = missing_val[1]/ unique_rows *100
null_prcnt_agent
````

````python
#Percentage of missing values in column company is less so it is better to fill those values with 0
df['agent'].fillna(value = 0, inplace = True)
df['agent'].isnull().sum() #rechecking column has missing values or not
````

````python
#percentage of null value in column country
null_prcnt_country = missing_val[2]/ unique_rows *100
null_prcnt_country
````

````python
#Percentage of missing values in column country is less so it is better to fill those values with 'others' as country name
df['country'].fillna(value = 'others', inplace = True)
df['country'].isnull().sum() #rechecking column has missing values or not
````

````python
#percentage of null value in column children
null_prcnt_children = missing_val[3]/ unique_rows *100
null_prcnt_children
````

````python
#Percentage of missing values in column children is less so it is better to fill those values with 0
df['children'].fillna(value = 0, inplace = True)
df['children'].isnull().sum() #rechecking column has missing values or not
````

````python
#checking whether the database having any other null value or not
df.isnull().sum()
````

#### Chart - 2: Average ADR of each Hotel type

````python
# Chart - 2 visualization code

# Group by Hotel
group_by_hotel = df.groupby('hotel')

# Grouping by Hotel adr
highest_adr = group_by_hotel['adr'].mean().reset_index()

# Set plot size
plt.figure(figsize = (8,4))

# Create the figure object
ax = sns.barplot(x= highest_adr['hotel'], y= highest_adr['adr'])

# Set labels
ax.set_xlabel("Hotel type", fontsize = 20)
ax.set_ylabel("ADR", fontsize = 20)
ax.set_xticklabels(['City Hotel', 'Resort Hotel'], fontsize = 16)
ax.set_title('Average ADR of each Hotel type', fontsize = 20)

# To show
plt.show(ax)
````

##### 1. Why did you pick the specific chart?

Bar charts visually represent the distribution of values across various categories of a categorical or nominal variable, showcasing frequency counts. Occasionally, these charts also display additional statistics like percentages.

In this instance, I utilized a Bar chart to effectively illustrate the average ADR for each type of hotel. This method offers a clear and practical visualization of the data.

The City hotel boasts the highest ADR, indicating that City Hotels are generating more revenue compared to Resort Hotels. The correlation is straightforward: a higher ADR corresponds to increased revenue.

````python
# Chart - 8 visualization code

# Using groupby on arrival_date_month and taking the hotel count
bookings_by_months_df = df.groupby(['arrival_date_month'])['hotel'].count().reset_index().rename(columns = {'hotel':'Counts'})

# Creating list of months in order
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']

# Creating dataframe which will map the order of above months list without changing its values
bookings_by_months_df['arrival_date_month'] = pd.Categorical(bookings_by_months_df['arrival_date_month'], categories = months, ordered = True)

# Sorting by arrival_date_month
bookings_by_months_df = bookings_by_months_df.sort_values('arrival_date_month')

bookings_by_months_df
````

````python
# Visualizing with the help of line plot

# Set plot size
plt.figure(figsize = (10,4))

# Plotting lineplot on x- months & y- bookings counts
sns.lineplot(x = bookings_by_months_df['arrival_date_month'], y = bookings_by_months_df['Counts'])

# Set title
plt.title('Number of bookings across each month', fontsize = 20)

# Set labels
plt.xlabel('Month', fontsize = 16)
plt.ylabel('Number of bookings', fontsize = 16)

# To show
plt.show()
````

##### 1. Why did you pick the specific chart?

I opted for a line chart because it's great at revealing small changes that might be missed in other graphs. It makes it easy to see trends over different months, providing a straightforward way to track the variation in the 'number of bookings.'
